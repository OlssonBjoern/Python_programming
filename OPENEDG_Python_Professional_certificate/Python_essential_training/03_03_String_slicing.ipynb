
# Strings

### Slicing

```python
name = 'My name is Tony Stark'
```

```python
name[0]
```

```python
name[1]
```

```python
# Gets the first 8 indexes in the string
name[0:7]
```

```python
# Can also do the other way around if you dont know the end of the string for example
name[11:]
```

### Formatting

```python
'My number is: ' + str(5)
```

```python
f'My number is {5}'
```

```python
# String formatting, putting in expressions with { }
f'My number is {5}, and twice that is {2*5}'
```

```python
f'Pi is: {math.pi:.2f}'
```
