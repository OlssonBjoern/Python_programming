# Class

```python
class Dog: 
    # If attributes like legs is moved outside of the init function like:
    # legs = 4, it becomes a static attribute
    def __init__(self, name):
        self.name = name
        self.legs = 4

    

    def speak(self):
        print(self.name + 'says: Bark!')

myDog = Dog('Fido')
print(myDog.name)
print(myDog.legs)
```
