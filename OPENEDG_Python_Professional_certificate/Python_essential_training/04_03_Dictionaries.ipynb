# Dictionaries

```python
footballers = {
    'goat' : 'Leo Messi',
    'constant second' : 'Cristiano Ronaldo',
    'swedish god': 'Zlatan Ibrahimovic'
}
footballers
```

```python
footballers['goat']
```

```python
footballers['Il Capitano'] = 'Paolo Maldini'
footballers.keys()
```

```python
footballers.values()
```

```python
# Cant manipulate the dictionary, however you can make a dictionary of lists like this:

animals = {
    'a' : ['anaconda', 'antelope'],
    'b' : ['bear'],
    'c' : ['chinchilla', 'cat']
}

# Accessing elements now become:

animals['b'].append('bison')

animals
```

```python
animals['d'] = ['dog']
animals
# This syntax will however possibly overwrite 'd' if it is already a KEY in the list
# What can be done instead is:
```

```python
if 'd' not in animals:
    animals['d'] = []

animals['d'].append('dog')

print(animals)
```

### Default dict

```python
# Import default dict from collections

animals = defaultdict(list)
```

```python
animals
```

```python
animals['e'].append('elephant')
animals
```

```python
animals['e'].append('elk')
animals['a']
animals
```
