## Control flow

```python
a = True
b = True
```

```python
if a:
    print('A is true!')
    print('Also print this')
    if b:
        print('Both a and b are TRUE')
else:
    print('a is FALSE!')

print('This prints no matter the if-statement')
```

### For loops

```python
ab = [1, 2, 3, 4, 5]

for item in ab:
    if ab[item] == 3:
        print('We found 3!!')
        break
```

### While loop

```python
i = 0
while i < 5:
    print(i)
    i += 1
```
