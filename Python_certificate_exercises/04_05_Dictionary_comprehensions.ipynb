# Dictionary comprehensions

```python
# List of animal tuples
animalList = [('a', 'anaconda'), ('b', 'bear'), ('c', 'cobra'), ('d', 'dolphin')]

animals = {item[0]: item[1] for item in animalList}
animals
```

```python
animals.items()
```

```python
# Turning the dictionary back into the list of tuples
list(animals.items())
```

```python
ironMan = 'Tony Stark'
for letter in ironMan:
    print(letter)
```
