# Files, reading, writing

```python
f = open('file_exercise.txt', 'r')
# This print returns a file object
print(f)
```

```python
# Read lines 1 by 1
f.readline()
```

```python
# To read easier we can do
f.readlines()
```

```python
f = open('file_exercise.txt', 'r')
for line in f.readlines():
    print(line.strip())
```

# Write to files

```python
f = open('file_output_exercise.txt', 'w')
print(f)
```

```python
f.write('Line 1')
f.write('Line 2')
```

```python
f.close()
```

```python
# To not overwrite the file each time, use append mode

f = open('file_output_exercise.txt', 'a')
f.write('Line 3')
f.write('Line 4')
f.close()
```

```python
with open('file_output_exercise.txt', 'a') as f:
    f.write('\nSome new stuff')
    f.write('\nSome EVEN NEWER stuff')
```
