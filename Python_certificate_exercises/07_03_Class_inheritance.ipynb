# Inheritance

```python
class Dog: 
    # If attributes like legs is moved outside of the init function like:
    # legs = 4, it becomes a static attribute
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(self.name + ' says: Bark!')

    def getLegs(self):
        return self._legs

class Heeler(Dog):
    def speak(self):
        print(f'{self.name} says: Rough! Rough!' )

    def wagTail(self):
        print('Wagging and causing shin bruised all around!')
```

```python
dog = Heeler('Zach')
dog.speak()
dog.wagTail()
```
