# Static and Instance Methods

```python
class WordSet:
    # Start off with an empty set that we will fill in
    def __init__(self):
        self.words = set()

    def addText(self, text):
        # 
        text = WordSet.cleanText(text)
        # Splits each string on spaces, so it turns a sentence into a list of words
        for word in text.split():
            self.words.add(word)

    def cleanText(text):
        # Chaining functions
        text = text.replace('!', '').replace('.', '').replace(',', '').replace('\'', '')
        return text.lower()

wordSet = WordSet()

wordSet.addText('Hi, I\'m King T\'Challa! Here is my favourite sentence: WAKANDA FOREVER!')
wordSet.addText('Here is another one: WAKANDA almost FOREVER!')

print(wordSet.words)
```
