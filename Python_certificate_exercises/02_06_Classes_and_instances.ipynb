# Classes

```python
class Dog:
        # init function to initialize the function, specially defined in Python
        def __init__(self, name):
            self.name = name
            self.legs = 4
        
        def speak(self):
            print(self.name + ' says: Bark!')
```

```python
# Initialize dog class
my_dog = Dog('Rover')
another_dog = Dog('Fluffy')
```

```python
my_dog.speak()
another_dog.speak()
```
