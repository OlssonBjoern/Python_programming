# Booleans again

```python
# Everything inside the bool function that isn't an empty string is True
bool(1)
```

```python
myList = [1, 2]
if myList:
    print('My list has some values in it')
# This ^ is equivalent to writing bool(myList), since this will equate to True
```

```python
weatherIsNice = True
haveUmbrella = False

# If not is the same as if condition == false
# Here the issue is that it evaluates haveUmbrella first which it considers to be true.
if not (haveUmbrella or weatherIsNice):
    print('Stay inside')
else:
    print('Go for a walk')
```
